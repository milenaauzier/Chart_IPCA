```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "C:/Users/milen/Desktop/Projetos/ipca.csv"
```

```python
df = pd.read_csv(path,sep = ';')
```

```python
or_df = df.sort_values(by='Inflacao', ascending = True)
```

```python
or_df
```

```python
plt.subplots(figsize=(17, 20))

plt.style.use('ggplot')
plt.hlines(y= or_df['Cidade'], xmin = 0, xmax = or_df['Inflacao'], color = 'teal')
plt.plot(or_df['Inflacao'], or_df['Cidade'], "o")


plt.title("VARIAO DA INFLAO MENSAL POR CIDADE (%) (MARO/2022)", loc='center', fontsize=20)
plt.xlabel('IPCA mensal', fontsize = 20)
plt.ylabel('Cidades', fontsize = 20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.savefig("IPCAggplot.png", dpi = 350, bbox_inches='tight')

plt.show()

```
